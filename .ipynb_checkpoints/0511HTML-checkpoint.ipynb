```python
from bs4 import BeautifulSoup #變成BeautifulSoup的html格式

path = "web.html" #路徑從notebook所在位置開始算
with open (path,"r",encoding="UTF-8") as f: #r是讀取,提供解碼方式
    soup=BeautifulSoup(f.read(),"html")

print(soup.head.title.text) #.title找title夾的中間
```

```python
print(soup.prettify())  #prettif會分階層比較好看
```

```python
print(soup.find_all("h1")) #find_all可省略,整理成一個list
print(soup.find("h1",class_="abc")) #取value
```

```python
print(soup.select(".abc")) #找abc的class #list
```

```python
print(soup.select("h1[class=job]")) #list
```

```python
print(soup.h2.text)
```

```python
#讀取104網站

import requests
from bs4 import BeautifulSoup

res=requests.get("https://www.104.com.tw/jobs/search/?keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&order=1&jobsource=2018indexpoc&ro=0")

job=BeautifulSoup(res.text,"html")
print(job.prettify())
```

```python
print(job.article)
```

```python
print(job.article["data-job-name"])
print(job.article["data-cust-name"])
print("https:"+job.article.find("a",class_="js-job-link")["href"])
print(job.article("li")[2].text)
print(job.article("a")[2].text)
```

```python
n=0

for work in job("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item js-job-item--focus b-block--ad"):
    print(work["data-job-name"])
    print(work["data-cust-name"])
    print("https:"+work.find("a",class_="js-job-link")["href"])
    print(work("li")[2].text)
    print(work("a")[2].text)
    n+=1
    print(n)
```
