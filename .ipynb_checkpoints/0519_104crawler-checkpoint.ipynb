```python
import requests
from bs4 import BeautifulSoup
from time import sleep


res = requests.get("https://www.104.com.tw/jobs/search/?keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&order=1&jobsource=2018indexpoc&ro=0")
soup = BeautifulSoup(res.text)

page = 1
while soup.find_all("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item") !=[]:
    print("now is page"+str(page))
    print("----------------------------------")
    for job in soup.find_all("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item"):
        print(job.a.text)
        print("http:"+job.a["href"])
        print(job.find("ul",class_="b-list-inline b-clearfix").a.text.strip()) #.strip()把資料空格清乾淨
        print(job.find("ul",class_="b-list-inline b-clearfix job-list-intro b-content").li.text)
        if job.find("div",class_="job-list-tag b-content").select("span") ==[]:
            print(job.find("div",class_="job-list-tag b-content").a.text)
        else:
            print(job.find("div",class_="job-list-tag b-content").span.text)
        print("----------------------------------")
    
    page+=1
    res = requests.get("https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1&page=+str(page)")
    soup = BeautifulSoup(res.text)
    sleep(2)
```

```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import openpyxl

wb = openpyxl.Workbook()
ws = wb.active

ws["A1"] = "職缺名稱"
ws["B1"] = "職缺連結"
ws["C1"] = "公司名稱"
ws["D1"] = "工作地區"
ws["E1"] = "薪資待遇"

res = requests.get("https://www.104.com.tw/jobs/search/?keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&order=1&jobsource=2018indexpoc&ro=0")
soup = BeautifulSoup(res.text)

page = 1
while soup.find_all("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item") !=[]:
#while len(soup.find_all("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item")) <50:
    for job in soup.find_all("article",class_="b-block--top-bord job-list-item b-clearfix js-job-item"):
        a= job.a.text
        b= "http:"+job.a["href"]
        c= job.find("ul",class_="b-list-inline b-clearfix").a.text.strip() #.strip()把資料空格清乾淨
        d= job.find("ul",class_="b-list-inline b-clearfix job-list-intro b-content").li.text
        if job.find("div",class_="job-list-tag b-content").select("span") ==[]:
            e= job.find("div",class_="job-list-tag b-content").a.text
        else:
            e= job.find("div",class_="job-list-tag b-content").span.text
    ws.append([a,b,c,d,e])
    
    page+=1
    res = requests.get("https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1&page=+str(page)")
    soup = BeautifulSoup(res.text)
    sleep(1)

    wb.save("1120522_104_bigdatajob.xlsx")
```
