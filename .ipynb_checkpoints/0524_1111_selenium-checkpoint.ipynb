```python
from selenium import webdriver #導入瀏覽器
from selenium.webdriver.common.by import By #導入選取器
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options #瀏覽器設定

#開啟瀏覽器
driver = webdriver.Chrome("C:\\Users\\user\\chromedriver.exe")
driver.get("https://www.1111.com.tw/")

search = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR,"input[class='UISearchbarInput ui-autocomplete-input']")))
search.send_keys("大數據")

go = driver.find_element(By.CSS_SELECTOR,"a[class='btn_gradient btn_size_1 UI_searchbar_btn']")
go.click()
```

```python
from bs4 import BeautifulSoup
from time import sleep

soup = BeautifulSoup(driver.page_source,"html.parser")
print(soup)
```

```python
print(soup.find_all("div",class_="job_item_info"))
```

```python
n=0
for job in soup.find_all("div",class_="job_item_info"):
    n+=1
    print(n)
    print(job.h5.text)
    print(job.a["href"])
    print(job.h6.text)
    print(job.find("a",class_="job_item_detail_location mr-3 position-relative").text)
    print(job.find("div",class_="job_item_detail_salary ml-3 font-weight-style digit_6").text)
    print("--------")
```

```python
scroll = 0
for time in range(1,3):
    scroll+=1
    driver.execute_script("window.scrollTo(0,10000)")
    print(scroll)
    n=0
    for job in soup.find_all("div",class_="job_item_info"):
        n+=1
        print(n)
        print(job.h5.text)
        print(job.a["href"])
        print(job.h6.text)
        print(job.find("a",class_="job_item_detail_location mr-3 position-relative").text)
        print(job.find("div",class_="job_item_detail_salary ml-3 font-weight-style digit_6").text)
        print("--------")
```

```python
from selenium.webdriver.common.keys import Keys #模擬按鍵
#對driver執行動作，送出enter鍵，執行()
webdriver.ActionChains(driver).send_keys(Keys.ENTER).perform()

#對driver執行script語法(將螢幕滑動到(x,y))
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```
