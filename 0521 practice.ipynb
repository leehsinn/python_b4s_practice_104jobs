```python
import requests
from bs4 import BeautifulSoup

res=requests.get("https://www.organicshops.cc/categories.aspx?cid=235")

soup=BeautifulSoup(res.text)
print(soup)
```

```python
n=0
for items in soup.find_all("div",class_="boxitem"):
    n+=1
    print(n)
    print(items.h4.text)
    print(items.span.text)

#print(soup.find_all("div",class_="boxitem"))
```

```python
for drink in soup.find_all("h4",class_="itemtitle"):
    print(drink.text)
```
